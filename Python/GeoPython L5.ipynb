```python
#store file path as separate variable
fpath = r"C:/Users/PC COMPUTER/Desktop/JG/SCIENCE/BIOFORCLIM_DR/stacje_meteo/BabiaG/_DANE/2022_01_07/Lukasz_babiagora_3_DANE_Plaj_2022_01_07.dat"

#read .dat file to pandas
data = pd.read_table(fpath,
              sep = ",", #set column separator
              skiprows = [0, 1, 2, 3], #skip first four rows
              names = ['TIMESTAMP', 'RECORD', 'BAT', 'TCP', 'TC', 'RH', 'WD', 'WS']) #set column names

#convert columns from object (string) to float
data['TC'] = data['TC'].astype(float)
data['RH'] = data['RH'].astype(float)
```
