```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from windrose import WindroseAxes
```

```python
station = 'Mszana_Dolna'
station
```

```python
wd_data = pd.read_csv('D:/JG/SCIENCE/BIOFORCLIM_DR/Meteo_wind_regime/data/imgw/telem/wd/'+station+'.csv')
wd_data['wd'] = pd.to_numeric(wd_data['wd'], errors = 'coerce')
wd_data = wd_data[wd_data.wd <= 360]
wd_data
```

```python
print(wd_data['wd'].max()) #check max value of wind direction
n = len(wd_data)
print(n)
```

```python
#the rose of aspect directions of monitoring plots - windthrows
wd = list(wd_data['wd']) #wind direction data
ws = list(0 for i in range(len(wd))) #speed data - list of 0s
ax = WindroseAxes.from_ax() #plot windrose
ax.bar(direction = wd, var = ws, edgecolor = 'white', color = 'steelblue') #plot windrose - reffer to data
ax.set_title(station+' - wind direction 2008 - 2022', size = 15, weight = 'bold') #set title
ax.set_xlabel(f'number of observations: {n}', size = 12) #set xaxis label

os.chdir(r'D:\\JG\SCIENCE\\BIOFORCLIM_DR\\Meteo_wind_regime\\plots') #set directory
plt.savefig(station+'_wd.png', dpi = 180) #save plot

plt.show()
```
